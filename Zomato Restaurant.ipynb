# Data Storage:

This problem statement contains two datasets- **Zomato.csv** and **country_code.csv**.
**Country_code.csv** contains two variables:

• Country code

• Country name


The collected data has been stored in the Comma Separated Value file Zomato.csv. Each
restaurant in the dataset is uniquely identified by its Restaurant Id. Every Restaurant contains the following variables:

• Restaurant Id: Unique id of every restaurant across various cities of the world

• Restaurant Name: Name of the restaurant

• Country Code: Country in which restaurant is located

• City: City in which restaurant is located

• Address: Address of the restaurant

• Locality: Location in the city

• Locality Verbose: Detailed description of the locality

• Longitude: Longitude coordinate of the restaurant&#39;s location

• Latitude: Latitude coordinate of the restaurant&#39;s location

• Cuisines: Cuisines offered by the restaurant

• Average Cost for two: Cost for two people in different currencies ��

• Currency: Currency of the country

• Has Table booking: yes/no

• Has Online delivery: yes/ no
    
• Is delivering: yes/ no

• Switch to order menu: yes/no

• Price range: range of price of food

• Aggregate Rating: Average rating out of 5

• Rating color: depending upon the average rating color

• Rating text: text on the basis of rating of rating

• Votes: Number of ratings casted by people

Problem statement : In this dataset predict 2 things –

1) Average Cost for two

2) Price range

# Reducing Skewness

```python
#Trying to reduce skewness using cube root method
a=new_df["Votes"]
b=new_df["Longitude"]
c=new_df["Latitude"]




a=np.cbrt(a)
b=np.cbrt(b)
c=np.cbrt(c)





print("Votes skewness:",a.skew())
print("Longitude skewness:",b.skew())
print("Latitude skewness:",c.skew())
```

**Here you can see cube root can't reduce all the columns to our acceptable skewness range.**

```python
#Trying to reduce skewness using square root method

a=new_df["Votes"]
b=new_df["Longitude"]
c=new_df["Latitude"]




a=np.sqrt(a)
b=np.sqrt(b)
c=np.sqrt(c)





print("Votes skewness:",a.skew())
print("Longitude skewness:",b.skew())
print("Latitude skewness:",c.skew())
```

**Not working.**

```python
#Trying to reduce skewness using log method


a=new_df["Votes"]
b=new_df["Longitude"]
c=new_df["Latitude"]




a=np.log(a)
b=np.log(b)
c=np.log(c)





print("Votes skewness:",a.skew())
print("Longitude skewness:",b.skew())
print("Latitude skewness:",c.skew())
```

**Not working.**

```python
#Trying to reduce skewness using log1p method
a=new_df["Votes"]
b=new_df["Longitude"]
c=new_df["Latitude"]




a=np.log1p(a)
b=np.log1p(b)
c=np.log1p(c)





print("Votes skewness:",a.skew())
print("Longitude skewness:",b.skew())
print("Latitude skewness:",c.skew())
```

**Here you can see Votes column's skewness can be reduced by using log1p.So I will implement it permanently on this column.**

```python
#Implementing log1p on Votes column
new_df["Votes"]=np.log1p(new_df["Votes"])
```

```python
#Checking the skewness again
new_df.skew()
```

**Now you can see Votes range column's skewness is in my acceptable range.**

```python
#Trying to reduce skewness using PowerTransform method 


b=new_df["Longitude"]
c=new_df["Latitude"]


#using PowerTransformer to reduce the skewness of CS-106 ,HS-304,CS-302,CS-419 these columns
from sklearn.preprocessing import PowerTransformer

pt=PowerTransformer("yeo-johnson")   #yeo-johnson method works on both positive and Negative method


b_reshape=b.values.reshape(-1,1)
c_reshape=c.values.reshape(-1,1)



#Now fitting the Transformer

b_transform=pt.fit_transform(b_reshape)
c_transform=pt.fit_transform(c_reshape)


#Converting transformed array back to the DataFrame

b=pd.DataFrame(b_transform)
c=pd.DataFrame(c_transform)



print("Longitude skewness",b.skew())
print("Latitude skewness",c.skew())
```

**Here you can see Latitude column's skewness can be reduced by using PowerTransformer,So I will implement it permanently.**

```python
#Implementing PowerTransformer on  Latitude column

from sklearn.preprocessing import PowerTransformer

pt=PowerTransformer("yeo-johnson")   #yeo-johnson method works on both positive and Negative method

Latitude_reshape=new_df["Latitude"].values.reshape(-1,1)


#Now fitting the Transformer
Latitude_transform=pt.fit_transform(Latitude_reshape)


#Converting transformed array back to the DataFrame
new_df["Latitude"]=pd.DataFrame(Latitude_transform)
```

# Finding the best Random State

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error

MaxAcc=0
MaxRs=0

for i in range(1,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=i)
    RFR=RandomForestRegressor()
    RFR.fit(x_train,y_train)
    pred=RFR.predict(x_test)
    Acc=r2_score(y_test,pred)
    
    if Acc>MaxAcc:
        MaxAcc=Acc
        MaxRs=i
        
print("Max R2 Score is",MaxAcc,"in the random state: ",MaxRs)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
MaxAcc=0
MaxRs=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.30,random_state=i)
    RFC=RandomForestClassifier()
    RFC.fit(x_train,y_train)
    pred=RFC.predict(x_test)
    acc=accuracy_score(y_test,pred)
    if acc>MaxAcc:
        MaxAcc=acc
        MaxRs=i
        
print("The best accuracy score is",MaxAcc,"at Random State",MaxRs)
```
