# Insurance Claim Fraud Detection

# Project Description

Insurance fraud is a huge problem in the industry. It's difficult to identify fraud claims. Machine Learning is in a unique position to help the Auto Insurance industry with this problem.

In this project, you are provided a dataset which has the details of the insurance policy along with the customer details. It also has the details of the accident on the basis of which the claims have been made. 

In this example, you will be working with some auto insurance data to demonstrate how you can create a predictive model that predicts if an insurance claim is fraudulent or not.

# Reducing Skewness

```python
#Trying to reduce skewness using cube root method
a=new_df["vehicle_claim"]
b=new_df["total_claim_amount"]


a=np.cbrt(a)
b=np.cbrt(b)



print("vehicle_claim skewness:",a.skew())
print("total_claim_amount skewness:",b.skew())
```

**So here you can see cube root method is not working.**

```python
#Trying to reduce skewness using square root method
a=new_df["vehicle_claim"]
b=new_df["total_claim_amount"]


a=np.sqrt(a)
b=np.sqrt(b)



print("vehicle_claim skewness:",a.skew())
print("total_claim_amount skewness:",b.skew())
```

**sqrt is also not working.**

```python
#Trying to reduce skewness using log method
a=new_df["vehicle_claim"]
b=new_df["total_claim_amount"]


a=np.log(a)
b=np.log(b)



print("vehicle_claim skewness:",a.skew())
print("total_claim_amount skewness:",b.skew())
```

```python
#Trying to reduce skewness using log1p method
a=new_df["vehicle_claim"]
b=new_df["total_claim_amount"]


a=np.log1p(a)
b=np.log1p(b)



print("vehicle_claim skewness:",a.skew())
print("total_claim_amount skewness:",b.skew())
```

**Still not working**

```python
from sklearn.preprocessing import PowerTransformer
a=new_df["vehicle_claim"]
b=new_df["total_claim_amount"]

reshape_a=a.values.reshape(-1,1)
reshape_b=b.values.reshape(-1,1)

pt=PowerTransformer("yeo-johnson")  #yeo-johnson method works on both positive and Negative method

#Now fitting the Transfomer
a_transformed=pt.fit_transform(reshape_a)
b_transformed=pt.fit_transform(reshape_b)

#Convert the transformed array back to the DataFrame
a=pd.DataFrame(a_transformed)
b=pd.DataFrame(b_transformed)


print("vehicle_claim skewness:",a.skew())
print("total_claim_amount skewness:",b.skew())
```

**Now you can see finally I can reduce these columns skewness by using the PowerTransformer.So I will implement this techenique permanently on these columns**

# Modeling

# Finding the best Random State

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
MaxAcc=0
MaxRs=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.30,random_state=i)
    RFC=RandomForestClassifier()
    RFC.fit(x_train,y_train)
    pred=RFC.predict(x_test)
    acc=accuracy_score(y_test,pred)
    if acc>MaxAcc:
        MaxAcc=acc
        MaxRs=i
        
print("The best accuracy score is",MaxAcc,"at Random State",MaxRs)
```
